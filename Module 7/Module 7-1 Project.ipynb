## My data analysis of movie review data

### Question 1: Which critic has written the most reviews?
### Question 2: Which is the most-reviewed movie?

#### First I import pandas as pd

```python
import pandas as pd
```

#### Next I read my .csv file into pandas

```python
df = pd.read_csv('reviews.csv')
df.head() # Here I am printing just the first 5 rows to get a look at my data frame and how it is organized.
```

#### I'd like to get a sense of how large my dataset is, so I am going to use df.shape to see the shape of my dataframe

```python
df.shape
```

#### That's a lot of rows!

#### I suspect that there are probably some duplicate entries in this dataset. I am going to see what happens when I call the drop_duplicates( ) method.

```python
df.drop_duplicates(subset='quote',keep='first',inplace=True)
df.head()
```

```python
df.shape
```

#### So, deleting duplicates did not have a huge impact on the dataset, but it may have eliminated some noise from my data.

### In order to answer my first question, I need to return the frequency with which each critic's name appears in the "critic" column. For this I call the _describe( )_ function, which will print me the "top," or most frequently occurring value, as well as its frequency. 

```python
df['critic'].describe()
```

#### So, Roger Ebert is the critic who has written the highest number of reviews, with 1101 of them.

### My next question asks which movie received the most reviews, so I'll use the same function with a different argument.

```python
df['title'].describe()
```

#### The most-reviewed movie in this dataset is _Beauty and the Beast_, with 20 reviews. Now can I see whether it justly received all "fresh" reviews?

```python
df[df.title=='Beauty and the Beast']
```

#### And look at that! Not a single "rotten" review in sight.
