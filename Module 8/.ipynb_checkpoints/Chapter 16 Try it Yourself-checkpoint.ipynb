```python
# Use matplotlib to create a simple plot of the daily highs

import csv

from matplotlib import pyplot as plt

filename = 'sitka_weather_07-2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    highs = []
    for row in reader:
        high = int(row[1])
        highs.append(row[1])

fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(highs,c='red')

plt.title("Daily high temperatures, July 2014", fontsize=24)
plt.xlabel('', fontsize=16)
plt.ylabel('Temperature (F)', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
# Add dates to make graph more useful

import csv
from datetime import datetime

from matplotlib import pyplot as plt

filename = 'sitka_weather_07-2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, highs = [],[]
    for row in reader:
        current_date = datetime.strptime(row[0], "%Y-%m-%d")
        dates.append(current_date)
    
        high = int(row[1])
        highs.append(high)
    
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red')

title="Daily high temperatures, July 2014\nSitka, AK"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel('Temperature (F)', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
# Extract the low temperatures from the csv file and add them to the graph.

import csv
from datetime import datetime

from matplotlib import pyplot as plt

filename = 'sitka_weather_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[0], "%Y-%m-%d")
        dates.append(current_date)
        
        high = int(row[1])
        highs.append(high)
        
        low = int(row[3])
        lows.append(low)

fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red')
plt.plot(dates, lows, c='blue')

title = "Daily high and low temperatures - 2014\nSitka, AK"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel('Temperature (F)', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
# Use shading to show the range between each day's high and low temperatures.

import csv
from datetime import datetime

from matplotlib import pyplot as plt

filename = 'sitka_weather_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[0], "%Y-%m-%d")
        dates.append(current_date)
        
        high = int(row[1])
        highs.append(high)
        
        low = int(row[3])
        lows.append(low)

fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red', alpha=0.5) # alpha argument controls color's transparency
plt.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

title = "Daily high and low temperatures - 2014\nSitka, AK"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel('Temperature (F)', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.ylim(10, 120)

plt.show()
```

### Exercise 16-2: Sitka-Death Valley Comparison
Change the settings for the y-axis on charts in fig. 16-5 and 16-6, or try plotting them on the same graph. I added the _y_lim()_ function to my two plots above to make sure they were using the same range, then wrote the function below to plot them in the same graph.

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

def get_weather_data(filename, dates, highs, lows):
    """Get the high and low temps from a data file."""
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
    
        for row in reader:
            try:
                current_date = datetime.strptime(row[0], "%Y-%m-%d")
                high = int(row[1])
                low = int(row[3])
            except ValueError:
                print(current_date, 'missing data')
            else:
                dates.append(current_date)
                highs.append(high)
                lows.append(low)

# Get data for Sitka
dates, highs, lows = [], [], []
get_weather_data('sitka_weather_2014.csv', dates, highs, lows)

# Plot Sitka weather
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red', alpha=0.6)
plt.plot(dates, lows, c='blue', alpha=0.6)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.15)

# Get data for Death Valley
dates, highs, lows = [], [], []
get_weather_data('death_valley_2014.csv', dates, highs, lows)

# Add Death Valley weather to plot
plt.plot(dates, highs, c='red', alpha=0.3)
plt.plot(dates, lows, c='blue', alpha=0.3)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.05)

# Format chart
title="Daily high and low temperatures - 2014"
title+= "\nSitka, AK and Death Valley, CA"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel('Temperature (F)', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.ylim(10, 120)

plt.show()
```

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

filename = 'sitka_weather_07-2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, rainfalls = [], []
    for row in reader:
            try:
                current_date = datetime.strptime(row[0], "%Y-%m-%d")
                rainfall = float(row[19]) # Rainfall data is in column 19
            except ValueError:
                print(current_date, 'missing data')
            else:
                dates.append(current_date)
                rainfalls.append(rainfall)
    

fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, rainfalls, c='blue', alpha=0.4)
plt.fill_between(dates, rainfalls, facecolor='blue', alpha=0.15)

plt.title("Daily Precipitation in Inches - July 2014\nSitka, AK", fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel('Precipitation (inches)', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

### Exercise 16-4: Explore
Generate more visualizations for any other weather aspect. I decided to stick with Sitka and look at the high and low humidity in July 2014, and then all of 2014.

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

filename = 'sitka_weather_07-2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, highs, lows = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            high = int(row[7])
            low = int(row[9])
        except ValueError:
            print(current_date, 'missing data') # Error-checking for missing data
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)

fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red', alpha=0.5)
plt.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

title="Daily high and low Humidity - July 2014\nSitka, AK"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel('Humidity (%)', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.ylim(10, 110)

plt.show()
```
