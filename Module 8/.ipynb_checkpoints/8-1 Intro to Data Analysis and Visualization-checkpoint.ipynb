```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x,y)
plt.title("Test Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x,y)
plt.plot(x, z)
plt.title("Test Plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x,y)
plt.plot(x, z)
plt.title("Test Plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
data = pd.read_csv('countries.csv')
data
```

```python
# Compare the population growth in the U.S. and China
```

```python
us = data[data.country == "United States"]
us
```

```python
china = data[data.country == "China"]
china
```

```python
# First plot the U.S. population by year
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
# Compare percentage growth from the first year 
```

```python
us.population
```

```python
# Divide each year's population value by the 1952 population to make the first value 1.
# Multiply by 100 to convert to percentage amount.
```

```python
us.population / us.population.iloc[0] * 100
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(['United States', 'China'])
plt.xlabel('Year')
plt.ylabel('Population Growth (first year = 100)')
plt.show()
```

### Produce a bar chart from __8-1_Multi.csv__ 

```python
import pandas as pd
%matplotlib inline 
```

```python
df = pd.read_csv('8-1_Multi.csv', index_col=0, thousands=',')
df
```

```python
p = df.plot(kind = 'bar');
p.set_title('2017 Sales', fontsize=24, color = 'purple');
p.set_ylabel('Units Sold', fontsize=12, color = 'purple');
p.set_xlabel('Month', fontsize=12, color = 'purple');
```

```python
df.plot(kind = 'bar', stacked=True)
```

```python
df.plot(kind = 'barh')
```

```python
df.plot(kind = 'barh', stacked=True)
```

```python
from matplotlib import pyplot as plt

Products = ['Cameras', 'Laptops', 'Printers', 'Desktops']
Sales = [19569, 30038, 20582, 24815]

actualFigure = plt.figure(figsize = (8,8))
actualFigure.suptitle("Total Sales by Product", fontsize=22)

pie = plt.pie(Sales, labels=Products, autopct='%.2f')
```
